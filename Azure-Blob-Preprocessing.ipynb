```python
trans = pd.read_csv("Dataset/transactions.csv")
trans = trans[trans['store_nbr'] == 44]
trans.drop(['store_nbr'],axis=1,inplace = True)
trans
```

```python
str44 = str44.merge(trans, how='left', on = 'date')
str44['transactions'].fillna(0.0, inplace = True)
str44
```

```python
holiday = pd.read_csv("Dataset/holidays_events.csv")
national_holi = holiday[holiday['locale'] == 'National']
national_holi
```

```python
str44 = str44.merge(national_holi, on='date', how='left')
```

```python
local_holi = holiday[(holiday['locale'] == 'Local') & (holiday['locale_name'] == 'Quito')]
local_holi['date'] = pd.to_datetime(local_holi['date'])
```

```python
str44['date'] = pd.to_datetime(str44['date'])
```

```python
for date in local_holi['date'].values:
    df = str44[str44['date'] == date]
    if (df.index.values):
        str44.loc[df.index]['type'] = local_holi['type'][local_holi['date'] == date]
        str44.loc[df.index]['locale'] = local_holi['locale'][local_holi['date'] == date]
        str44.loc[df.index]['description'] = local_holi['description'][local_holi['date'] == date]
        str44.loc[df.index]['transferred'] = local_holi['transferred'][local_holi['date'] == date]
```

```python
from sklearn import preprocessing
label = preprocessing.LabelEncoder()
label.fit(str44['type'].values)
np.save('type_classes.npy', label.classes_)
str44['type'] = label.transform(str44['type'].values)

types = pd.DataFrame()
types['timestamp'] = pd.to_datetime(str44['date']).dt.strftime('%Y-%m-%dT%H:%M:%SZ')
types['value'] = str44['type'].values
types.to_csv("Azure Blob/type.csv", index = False)
types
```

```python
label = preprocessing.LabelEncoder()
label.fit(str44['locale'].values)
np.save('locale_classes.npy', label.classes_)
str44['locale'] = label.transform(str44['locale'].values)

locale = pd.DataFrame()
locale['timestamp'] = pd.to_datetime(str44['date']).dt.strftime('%Y-%m-%dT%H:%M:%SZ')
locale['value'] = str44['locale'].values
locale.to_csv("Azure Blob/locale.csv", index = False)
locale
```

```python
label = preprocessing.LabelEncoder()
label.fit(str44['description'].values)
np.save('description_classes.npy', label.classes_)
str44['description'] = label.transform(str44['description'].values)

description = pd.DataFrame()
description['timestamp'] = pd.to_datetime(str44['date']).dt.strftime('%Y-%m-%dT%H:%M:%SZ')
description['value'] = str44['description'].values
description.to_csv("Azure Blob/description.csv", index = False)
description
```

```python
label = preprocessing.LabelEncoder()
label.fit(str44['locale_name'].values)
np.save('locale_name_classes.npy', label.classes_)
str44['locale_name'] = label.transform(str44['locale_name'].values)

locale_name = pd.DataFrame()
locale_name['timestamp'] = pd.to_datetime(str44['date']).dt.strftime('%Y-%m-%dT%H:%M:%SZ')
locale_name['value'] = str44['locale_name'].values
locale_name.to_csv("Azure Blob/locale_name.csv", index = False)
locale_name
```
